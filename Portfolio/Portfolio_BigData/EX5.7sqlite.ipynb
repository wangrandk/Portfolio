```python
# choose "CustomerID,ProductID" from a joint table of Orders and 'Order Details', where the ProductID is the same as what 'ALFKI' bought.
import sqlite3 as lite
import sys

con = lite.connect('northwind.db')
con.text_factory = str
with con:

    cur = con.cursor()    
    #cur.execute("SELECT CustomerID,ProductID FROM Orders NATURAL JOIN 'Order Details' WHERE ProductID IN (SELECT ProductID FROM Orders AS O INNER JOIN 'Order Details' AS OD ON O.OrderID = OD.OrderID WHERE O.CustomerID='ALFKI') ")
    #cur.execute("SELECT CustomerID,ProductID FROM Orders NATURAL JOIN 'Order Details' WHERE ProductID IN (SELECT ProductID FROM Orders NATURAL JOIN 'Order Details' WHERE CustomerID='ALFKI') ")
    cur.execute("SELECT CustomerID FROM Orders NATURAL JOIN 'Order Details' WHERE ProductID IN (SELECT ProductID FROM Orders NATURAL JOIN 'Order Details' WHERE CustomerID='ALFKI') ")
    
    con.commit()
    
    c_p_id = cur.fetchall()
    for row in c_p_id:
        print row
```

```python
# list who bought which product the most.
from collections import Counter
print Counter(c_p_id).most_common() 

# we see that 'VAFFE' bought 4 productID 77, 'HUNGO' has bought 4 product 28, and so on. 
```
